```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import numpy as np
import time
```

```python
#读取txt文件
with open('./target.txt','r',encoding='UTF-8') as f:
    text = f.read()
#创建词云
wc = WordCloud(width=400,height=250,max_font_size=50,background_color='white').generate(text=text)
wc.to_file('./target.png')#保存
# 显示词云
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
#计时器
def func(f):
    def inner(*args,**kwargs):
        start_time = time.time()
        f(*args,**kwargs)
        end_time=time.time()
        print('耗时%s秒'%(end_time-start_time))
    return inner
```

```python
#指定行列数，矩阵元素为自然数列转化为二维矩阵
#cols=800
#rows=1000
cols=800
rows=1000
total=cols*rows

print('numpy计算：')
m=(np.arange(total)+1).reshape(rows,-1)
n=(np.arange(total)+1).reshape(rows,-1)
print('原始矩阵')
#print(m)
#print(n)

print('矩阵加法')
@func
def np_add():
    mat_add=m + n
    #print(mat_add)
np_add()

print('矩阵减法')
@func
def np_sub():
    mat_sub=m - n
    #print(mat_sub)
np_sub()

print('矩阵乘法')
@func
def np_mul():
    mat_mul=np.multiply(m,n)
    #print(mat_mul)
np_mul()

print('矩阵除法')
@func
def np_div():
    mat_div=m / n
    #print(mat_div)
np_div()

print('矩阵转置')
@func
def np_T():
    mat_T=m.T
    #print(mat_T)
np_T()
```

```python
print('列表计算：')
list1 = [[row*cols+col+1 for col in range(cols)] for row in range(rows)]
list2=[[row*cols+col+1 for col in range(cols)] for row in range(rows)]
print('原始矩阵')
#print(list1)
#print(list2)

print('矩阵加法')
@func
def list_add():
    li_add=[[k+l for k,l in zip(i,j)] for i,j in zip(list1,list2)]
    #print(li_add)
list_add()

print('矩阵减法')
@func
def list_sub():
    li_sub=[[k-l for k,l in zip(i,j)] for i,j in zip(list1,list2)]
    #print(li_sub)
list_sub()

print('矩阵乘法')
@func
def list_mul():
    li_mul=[[k*l for k,l in zip(i,j)] for i,j in zip(list1,list2)]
    #print(li_mul)
list_mul()

print('矩阵除法')
@func
def list_div():
    li_div=[[k*l for k,l in zip(i,j)] for i,j in zip(list1,list2)]
    #print(li_div)
list_div()

print('矩阵转置')
@func
def list_T():
    li_T=[[j[i] for j in list1] for i in range(cols)]
    #print(li_T)
list_T()
```
