```python
import random
import re
```

```python
#Q1 随机生成一个包含100个数字的list，将其写入文件，然后从文件中读取出来文件内容，将其排序（升序）后写入另一个文件。
def write_file(file_name,l):
    f = open(file_name,'w')
    num = 0
    for i in l:
        num += 1
        if(num != 100):
            f.write(str(i)+' ')
        else:
            f.write(str(i))
    f.close()
    
#生成长度为100的随机数组
my_list = []
for i in range(100):
    my_list.append(random.randint(0,100))
print('before:',my_list)
#将数组写入文件f.txt
write_file('./f.txt',my_list)

#读出数组
with open('./f.txt','r') as f1:
    temp_list = f1.read()
read_list = temp_list.split(' ')

read_list = [int(i) for i in read_list]

#计数排序
def count_sort(l):
    min = 100
    max = 0
    # 取得最大值和最小值
    for x in l:
        if x < min:
            min = x
        if x > max:
            max = x
    # 创建数组C
    count = [0] * (max - min +1)
    for index in l:
        count[index - min] += 1
    index = 0
    # 填值
    for a in range(max - min+1):
        for c in range(count[a]):
            l[index] = a + min
            index += 1
    return l
list2 = count_sort(read_list)
print('after sort:',list2)
#写回f2.txt
write_file('./f2.txt',list2)
```

```python
#Q2给定一个文件，可以从本作业附件中下载。统计其中每个单词出现的次数，并对其进行排序（倒序）。 
#读取text.txt
text_list = []
with open('./text.txt','r',encoding='UTF-8') as f:
    text = f.read()
#print(text)#原文本

#清洗标点符号
punc=  '()\.,\!?":;\/-'
def removePunctuation(text):
    text = re.sub(r'[{}]+'.format(punc),' ',text)
    return text.strip()
text2 = removePunctuation(text)
#print(text2)#去掉标点符号的文本

#用空格和换行符分隔单词
word_list = re.split('[\n\s]',text2)
#print(word_list)#单词表

#统计词频
count_dict = {}
for word in word_list:
    if len(word) == 0:
        continue
    else:
        if word in count_dict.keys():
            count_dict[word]=count_dict[word]+1
        else:
            count_dict[word] = 1
count_list = sorted(count_dict.items(),key = lambda x:x[1],reverse = True)

print(count_list)
```
